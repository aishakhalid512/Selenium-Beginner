webdriver is a Python Module which is available in Selenium Package. So, We already installed Selenium package.
So we got the webdriver module. In the webdriver module we have all the classes and Methods. So, by using these
classes and methods we should be able to Automate.

So, as a first step , in order to access those classes and Methods in python. We have to import this webdriver
module from Selenium Package.


What is a webdriver?

- webdriver is one of the component in selenium
- webdriver is a module
- So in this module multiple classes are created already
- So if you want to work with the chrome browser we have to call the chrome() class
- for firefox browser , we call the firefox() class
- for Edge browser we call the Edge() class
- So, in these classes some predefined methods are there which let us interact with the browser

```python
from selenium import webdriver  #Now we can access all the classes and methods from the webdriver module
```

```python
#Now we want to Open the Web browser, we have to call a class in webdriver
#Since we chose chrome browser , we will automate all these tasks for Chrome browser
#we access chrome class from the webdriver module
#webdriver.Chrome()
#the Chrome class constructor will take the Browser driver as a Parameter
#When we create webdriver.Chrome() , we have to create an Object 
#driver = webdriver.Chrome()
#driver is an object (or simply a variable which stores the Chrome class of the webdriver)
#So, driver is the object of the Chrome() class
#We will slowly learn what all methods are available in the Chrome class
#So the Chrome() constructor expects one argument and that is the location of the Chrome browser driver (webdriver exe)
#So this way we intialize the Chrome driver
#This will launch our browser
#So, internally this Chrome() class  
#So, we launch our browser
#Hence we accomplished the first step : Open Web Browser (Chrome)

driver = webdriver.Chrome('C:\webdrivers\chromedriver.exe')

#Some people write an additional statement here , that is "executable path" , which is also correct
#driver = webdriver.Chrome(executable_path = 'C:\webdrivers\chromedriver.exe')
```

#### Now we work on the second step : Open URL https://admin-demo.nopcommerce.com/login

```python
#Now we have to launch the application URL
#So to perform all other operations we have certain methods which are available inside the Chrome() class
#So we will be able to access those methods through "driver" object that we created, 
#driver = webdriver.Chrome('C:\webdrivers\chromedriver.exe') 
#because "class" methods we cannot directly access
#Through these drivers we have to access multiple types of methods
#Now we will access the URL through the browser that we opened through 
#driver = webdriver.Chrome('C:\webdrivers\chromedriver.exe')
#for this we use driver.get() method
#in this we pass the URL of the application

driver.get("https://admin-demo.nopcommerce.com/login")
```

### Provide Email : admin@yourstore.com

Now we need to enter the user name in the browser:
    


But how are we going to enter the "user name" in the box
- How will our script identify this element , which is actually called a web element
- So, evereything thats there on the web page is called a web element
- Example : image , text box , button , link <--- All these are web elements
- So, the first thing is , our Automation Code (Python code) should identify the web element
- And then perform the action.
- What is the action here ? 
- We need to inspect the web page 
- Every web page is designed using HTML language
- So, if you just look at the source code of the web page , we can see certain number of attributes that are
available for those elements
- Right click on "inspect"
- As you keep hovering over the tags , the elements get highlighted



We can use one of the Attributes to identify this element
- "name" and "id" are supported by Selenium webdriver
- "type" is not supported 
- Once you find the element you can perform the action
- To find any element on the web page , you have certain methods
- And all the methods we have to access through driver object


- All these are methods
- Suppose I want to find an element by "name"

```python
#This below statement will help us find the element
#driver.find_element_by_name("txtUsername")
#After identifying the element , we have to pass the value into the input box 
#For this we will continue the statement
#driver.find_element_by_name("txtUsername").send_keys()
#send_keys() method will input the same value into the text box element

driver.find_element_by_name("Email").send_keys('admin@yourstore.com')
```

### Now we inspect the Password element

```python
#This is also an input element
#This time we use id to access the element

driver.find_element_by_id('Password').send_keys('admin')
```

### Now we need to click on the "Login" button

```python
from selenium.webdriver.common.by import By
```

```python
driver.find_element(By.CLASS_NAME, "buttons").click()
```

### Once you successfully log in verify the Title of the web page

```python
#We again need to Right Click and inspect the page
#Under the <head> tag we see the <title> tag
```

```python
driver.title
```

### Verify Title of the page : "Dashboard/ nopCommerce administration" (Expected)

```python
#So driver.title will give us the actual title of the web page
#we will store this in a variable

act_title = driver.title

#What is our expected title , the one we see under the title tag on the browser 

exp_title = "Dashboard / nopCommerce administration"

#WE need to validate if the actual title is equat to the expected title
#We use the if condition

if act_title == exp_title:
    print("Login test passed!")
else:
    print("Login test failed!")
    
driver.close()  #Close the browser
```
